# Representing the completed and canceled orders on a bar chart.

```python
order_status = ['completed','canceled']
order_counts = [205100,139428]
plt.bar(order_status,order_counts,color=['blue','red'])
plt.title('Number of Completed and Canceled Orders')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
for i in range(len(order_status)):
      plt.text(i,order_counts[i],str(order_counts[i]))
plt.show()
```

Question 5.
#what is the Total Order value from 2016 to 2018?

```python
f_df.columns
```

```python
order_value_table=f_df[['Yr', 'price', 'qty_ordered']]
order_value_table['value'] = order_value_table['price'] * order_value_table['qty_ordered']
order_value_table = order_value_table[order_value_table['Yr'] > 2015]
order_value_sum = order_value_table[['Yr', 'value']].groupby('Yr').sum()
order_value_sum.plot(kind ='bar')
plt.xlabel('YEAR')
plt.ylabel('TOTAL VALUE IN PAKISTAN RUPEE')
plt.title('TOTAL VALUE OR ORDERS OVER 3 YEARS')

```

In conclusion,the total value of orders in 2016 where low but in 2017 they increased however in 2018 the value of orders totally went down or decreased. there are many canceled orders which can lead to business colapse if management does not handle customers with canceled orders with great care.Follow ups are to be made and get reasons and causes for the cancelation and a detailed report should be made conclusively.
